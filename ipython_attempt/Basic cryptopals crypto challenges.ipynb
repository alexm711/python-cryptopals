```python
import codecs
def convertHex2Base64(Hex):
    return codecs.encode(codecs.decode(Hex, 'hex'), 'base64').decode()
```

```python
import binascii
import codecs
def fixed_xor(t1,t2): 
    t1 = [(int(a[i:i+2], 16)) for i in range(0, len(a), 2)]
    t2 = [(int(b[i:i+2], 16)) for i in range(0, len(b), 2)]
    byte_array = [(A ^ B) for A,B in zip(t1,t2)]
    return binascii.hexlify(bytearray(byte_array))
```
